```python
#Create bar chart
labels = ['2015', '2016', '2017', '2018']
new_sales = [56.5, 51.2, 58.1, 55.8]
used_sales = [9.6, 9.9, 10.4, 18.4]

x = np.arange(len(labels))  # the label locations
width = 0.45  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, new_sales, width, label='New')
rects2 = ax.bar(x + width/2, used_sales, width, label='Pre-owned')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Year')
ax.set_ylabel('Total $ Sales in Millions')
ax.set_title('2015-2018 Dealership Sales in California')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0, max(new_sales)+30)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.savefig("2015-2018_Dealership_Sales_in_California.png")
plt.savefig('sample.png', bbox_inches="tight")

plt.show()
```
